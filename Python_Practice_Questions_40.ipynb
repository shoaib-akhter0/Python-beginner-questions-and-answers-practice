### **Q40. Mini Project: Ask for Name, Age, City, and Course, store them in a dictionary, display the information, and print whether the student is eligible for an internship based on age.**

```python
Name = input("Enter your name: ")
Age = int(input("Enter your age: "))
City = input("Enter your city: ")
Course = input("Enter your course: ")

Dictionary = {
    "Name" : Name,
    "Age" : Age,
    "City" : City,
    "Course" : Course
}
print(Dictionary)

if Age >= 18:
  print("You are eligible for an internship")
else:
  print("You are not eligible for an internship")
```
